# Workspace Setup
- Define Catalog and Schema for reuse
- Use `%run ./workspaceSetup`

Use this notebook to configure your Catalog and Schema

```python
list_catalogs = []
list_schemas = []
my_dict = {}
# Get list of all the catalogs and schemas within each catalog
list_catalogs = spark.sql("SHOW CATALOGS").collect()

# Update my_dict with catalog and schema names
for catalogName in list_catalogs:
    currentCatalog = catalogName.catalog
    spark.sql(f"USE CATALOG {currentCatalog}")
    list_schemas = spark.sql("SHOW SCHEMAS").collect()
    my_dict[currentCatalog] = []
    for schemaName in list_schemas:
        # Append databaseName to the dictionary
        my_dict[currentCatalog].append(schemaName.databaseName)
```

```python
# View your catalog and schema list
display(list(my_dict.keys()))
display(list(my_dict['system']))
```

```python
# Add a widget to select catalog and schema
dbutils.widgets.removeAll()
dbutils.widgets.dropdown('catalog_selected', list(my_dict.keys())[0], list(my_dict.keys()), "Select Catalog")
dbutils.widgets.dropdown('schema_selected', list(my_dict.values())[0][0], list(my_dict[dbutils.widgets.get("catalog_selected")]), "Select Schema")
```

```python
# Initialize Catalog and Schema
catalogSelected = dbutils.widgets.get("catalog_selected")
schemaSelected = dbutils.widgets.get("schema_selected")
#print(catalogSelected)
#print(schemaSelected)

spark.sql(f"USE CATALOG {catalogSelected}")
spark.sql(f"USE SCHEMA {schemaSelected}")
```
