#1. GROUP BY

##a. Group rows based on grouping experessions
- GROUPING SETS
  - Explicitly list all the grouping columns as a list in GROUP BY GROUPING SETS (...)
- ROLLUP
  - Provides aggregations (total and sub-totals) by hierarchy of columns in GROUP BY clause.
- CUBE
  - Provides aggregations (total and sub-totals) by all possible combination of columns in GROUP BY clause.

GROUPING SETS

```sql
SELECT product, customerID, sum(quantity) AS total_quantity, sum(totalPrice) AS total_revenue FROM sales_transactions
GROUP BY GROUPING SETS ((product, customerID), (product), ())
ORDER BY total_revenue DESC
LIMIT 10
```

ROLLUP

```sql
SELECT product, customerID, sum(quantity) AS total_quantity, sum(totalPrice) AS total_revenue FROM sales_transactions
GROUP BY ROLLUP(product, customerID)
ORDER BY total_revenue DESC
LIMIT 10
```

CUBE

```sql
SELECT product, customerID, sum(quantity) AS total_quantity, sum(totalPrice) AS total_revenue FROM sales_transactions
WHERE product LIKE '%Almond%'
GROUP BY CUBE(product, customerID)
ORDER BY total_revenue, total_quantity DESC
LIMIT 10
```

##b. Group rows based on aggregate functions
- MIN, MAX, COUNT, SUM, AVG
- DISTINCT, FILTER

These aggregate functions should be used in the SELECT....FROM section

```sql
SELECT 
  product,
  customerID,
  MAX(quantity) AS max_quantity,
  MIN(quantity) AS min_quantity
FROM sales_transactions
-- WHERE customerID = 2000003
GROUP BY product, customerID
ORDER BY customerID, product ASC
```

```sql
SELECT 
  product, 
  customerID, 
  SUM(quantity) AS total_quantity,
  SUM(DISTINCT quantity) AS total_distinct_quantity
FROM sales_transactions
GROUP BY product, customerID
ORDER BY product, customerID
```

```sql
SELECT product, customerID, sum(quantity) FILTER( WHERE product LIKE '%Almond%') AS total_quantity, sum(totalPrice) AS total_revenue FROM sales_transactions
GROUP BY product, customerID
ORDER BY product, total_revenue DESC
LIMIT 10
```
